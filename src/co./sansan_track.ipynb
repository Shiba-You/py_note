```python
from collections import deque
```

```python
n, m = list(map(int, input().split()))
maze = [list(input()) for i in range(n)] #複数行の数値の入力を取得
```

```python
del sy, gy
for val, i in enumerate(maze):
  print(val, i)
  if "S" in i:
    sy = val
    sx = i.index("S")
  if "G" in i:
    gy = val
    gx = i.index("G")
  if "sy" in locals() and "gy" in locals():
    break
print((sx, sy), (gx, gy))
```

```python
# たどり着くまでにかかった回数をいれる多重リスト。
d = [[float("inf")] * m for i in range(n)]
# 1ノード降りるときにforで回せるようリスト化しておく
dx = [1, 0, -1, 0]
dy = [0, 1, 0, -1]
sb = 0
print(d)
```

```python
# 接したポイントを取り込む用のqueを定義
que = deque([])
# スタート地点をqueに加える。探索の際はqueから取り出すループを回すので最初だけ先に入れておく
que.append((sy,sx,sb))
# スタートからスタートにかかる距離は0なのでinfを0にかえる。
d[sy][sx] = 0

print(que)
print(d)
```

```python
# queがあるかぎり続ける
bs = 0
while que:
  # queから座標を取り出す。
  p = que.popleft()
  # 取り出した点がゴールなら終了
  if p[0] == gy and p[1] == gx:
    break
  #　それ以外の時は，接している点を4方向順に取得。一つずつ処理する。
  for i in range(4):
    nx = p[0] + dx[i]
    ny = p[1] + dy[i]
    nb = p[2]
    print("(p: {}, que: {}, nx: {}, ny: {}".format(p, que, nx, ny))
    # 新しく取得した座標に対しての処理
    # 迷路内に存在する座標であり，壁ではなく，まだ通ったことがなくdがinfの点なら処理する = 新しく取得した道候補の1つ
    if 0 <= nx < n and 0 <= ny < m and maze[nx][ny] != "#" and d[nx][ny] > d[p[0]][p[1]]+1:
      # 道候補をqueに加える。
      que.append((nx, ny, nb))
      # その道候補への経路までにかかった経路をdに加える。
      d[nx][ny] = d[p[0]][p[1]] + 1
    elif 0 <= nx < n and 0 <= ny < m and maze[nx][ny] == "#" and d[nx][ny] > d[p[0]][p[1]]+1+nb:
      # 道候補をqueに加える。
      d[nx][ny] = d[p[0]][p[1]] + 1 + nb
      nb += 1
      que.append((nx, ny, nb))
      # その道候補への経路までにかかった経路をdに加える。
```
