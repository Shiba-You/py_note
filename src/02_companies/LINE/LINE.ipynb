あなたはn日前に戻ることができるタイムマシンを発明し，過去に戻った．
したがって．今日を含む未来nにちかんの株式価格はわかっている．
これからn日にわたって最適な取引を行うことで，お金を儲けたい．
今日を一日目としたとき i (1 ≤ i ≤ n)日目の株価は a_i 万円である．
今日の朝の所持金は s 万円である．
一日に一回，その日の株価に応じてどれかを行うことができる．

* 何もしない
* 所持金を下回らない限りで，株を1株以上k株以下買う
* 所持している株の数を下回らない範囲で，株を1株以上k株以下売る

ただし，株の売買の最小単位は1株であるため，売買株数は非不整数である．
また一日に購入できる株の上限は k 株とする．
n 日間で利益を最大化した時，最終日の夜には所持金はいくらになっているか？
