```python
from collections import deque

def init():
  N, M = map(int, input().split())
  grid = []
  D_loc = []
  for i in range(N):
    row = list(input())
    for j, el in enumerate(row):
      if el == "D":
        # direction = ["E", "N", "W", "S"]
        D_loc.append([i, j, 0, 0]) # [y, x, t, direction]
        row[j] = "."
      elif el == "H":
        sx, sy = j, i
        row[j] = "."
    grid.append(row)
  print(N, M, grid, D_loc, sx, sy)
  # N, M, grid, D_loc, sx, sy = 3, 4, [['.', '.', '.', '.'], ['.', '.', '.', '.'], ['.', '.', '.', '.']], [[1, 1, 0, 0], [2, 3, 0, 0]], 0, 0
  return N, M, grid, D_loc, sx, sy

def move_D(N, M, grid, D_loc, ct):
  for idx, D in enumerate(D_loc):
    cy, cx, t, d = D
    if ct == t:
      break
    if d == 0:
      dy, dx = 0, 1
    elif d == 1:
      dy, dx = -1, 0
    elif d == 2:
      dy, dx = 0, -1
    elif d == 3:
      dy, dx = 1, 0
    if (0<=cy+dy<N) and (0<=cx+dx<M) and (grid[cy+dy][cx+dx]!="#"):
      D_loc[idx] = [cy+dy, cx+dx, t+1, d]
    else:
      d += 1
      d %= 4
      D_loc[idx] = [cy, cx, t+1, d]
  return D_loc

def judge_catch(D_loc, cy, cx):
  for idx, D in enumerate(D_loc):
    print("(cy, cx) : ({},{}), D: ({}, {})".format(cy, cx, D[0], D[1], cy == D[0], cx == D[1]))
    if cy == D[0] and cx == D[1]:
      goal = D_loc.pop(idx)
      return D_loc, goal, True # [y, x, t, direction]
  return D_loc, [0,0,-1,0], False

def bfs(N, M, grid, D_loc, sx, sy, st):
  q = deque([(sy, sx, st, 0)])
  while q:
    cy, cx, t, stay = q.popleft()
    ct = t+1
    flag = False

    # Dの移動
    D_loc = move_D(N, M, grid, D_loc, ct)

    # Hの移動
    for dy, dx in [[1,0],[-1,0],[0,1],[0,-1]]:
      ty, tx = cy+dy, cx+dx
      if (0<=ty<N) and (0<=tx<M) and (grid[ty][tx]=="."):
        grid[ty][tx] = ct
        q.append((ty, tx, ct, 0))
        # 捕まえたか調査
        D_loc, goal, flag = judge_catch(D_loc, ty, tx)
        if flag:
          return D_loc, goal
    grid[cy][cx] = ct
    stay += 1
    if stay <= (N*M)//2:
      q.append((cy, cx, ct, stay))
      D_loc, goal, flag = judge_catch(D_loc, cy, cx)
      if flag:
        return D_loc, goal
    else:
      return [], [0,0,-1,0]


def main():
  N, M, grid, D_loc, sx, sy = init()
  st = 0
  while D_loc:
    D_loc, start = bfs(N, M, grid, D_loc, sx, sy, st)
    sy, sx, st, _ = start
    if st==-1:
      return st
  return st

if __name__ == "__main__":
  ans = main()
  print(ans)
```
