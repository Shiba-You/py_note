```python
from collections import deque

def init():
  h , w  = map(int, input().split())
  sy, sx = map(int, input().split())
  gy, gx = map(int, input().split())
  grid = [list(input()) for _ in range(h)]
  return h-1, w-1, sy-1, sx-1, gy-1, gx-1, grid

def bfs():
  while q:
    cy, cx, t = q.popleft()
    if cy == gy and cx == gx:
      print(t)
      return 
    if (0<=cy+1<h) and (grid[cy+1][cx]=="."):
      grid[cy+1][cx] = "#"
      q.append((cy+1, cx, t+1))
    if (0<=cy-1<h) and (grid[cy-1][cx]=="."):
      grid[cy-1][cx] = "#"
      q.append((cy-1, cx, t+1))
    if (0<=cx+1<w) and (grid[cy][cx+1]=="."):
      grid[cy][cx+1] = "#"
      q.append((cy, cx+1, t+1))
    if (0<=cx-1<w) and (grid[cy][cx-1]=="."):
      grid[cy][cx-1] = "#"
      q.append((cy, cx-1, t+1))
  return 

if __name__ == "__main__":
  h, w, sy, sx, gy, gx, grid = init()
  q = deque([(sy, sx, 0)])
  bfs()
```
