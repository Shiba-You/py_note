サイコロを N 回振ったとき、出た目の積が D の倍数となる確率を求めよ。

https://atcoder.jp/contests/tdpc/tasks/tdpc_dice

Ex1:
Int : 2 6
Out : 0.416666667

Ex2:
Int : 3 2
Out : 0.875000000

```python
def init():
  n, d = map(int, input().split())
  return n, d

def dp():
  n, d = init()
  q2 = 0
  q3 = 0
  q5 = 0

  while d % 2 == 0:
    q2 += 1
    d //= 2
  while d % 3 == 0:
    q3 += 1
    d //= 3
  while d % 5 == 0:
    q5 += 1
    d //= 5
  if d != 1:
    print(0)
    exit()
  
  grid = [[[[0 for _ in range(q5+1)] for _ in range(q3+1)] for _ in range(q2+1)] for _ in range(n+1)]
  grid[0][0][0][0] = 1

  for i in range(n):
    for c2 in range(q2+1):
      for c3 in range(q3+1):
        for c5 in range(q5+1):
          p = grid[i][c2][c3][c5] / 6
          if p == 0:
            continue
          grid[i+1][c2][c3][c5] += p                          # 1
          grid[i+1][min(c2+1, q2)][c3][c5] += p               # 2
          grid[i+1][c2][min(c3+1, q3)][c5] += p               # 3
          grid[i+1][min(c2+2, q2)][c3][c5] += p               # 4
          grid[i+1][c2][c3][min(c5+1, q5)] += p               # 5
          grid[i+1][min(c2+1, q2)][min(c3+1, q3)][c5] += p    # 6

  return grid[n][c2][c3][c5]

if __name__ == "__main__":
  ans = dp()
  print(ans)
```

```python
N, D = map(int, input().split())
a = 0
b = 0
c = 0
while D % 2 == 0:
  D //= 2
  a += 1
while D % 3 == 0:
  D //= 3
  b += 1
while D % 5 == 0:
  D //= 5
  c += 1
if D > 1:
  print(0)
  exit()

dp = [[[[0 for _ in range(c + 1)] for _ in range(b + 1)] for _ in range(a + 1)] for _ in range(N + 1)]
dp[0][0][0][0] = 1
for n in range(N):
  for i in range(a + 1):
    for j in range(b + 1):
      for k in range(c + 1):
        p = dp[n][i][j][k] / 6
        if p == 0:
          continue
        print("n : ", n)
        print(*dp, sep="\n")
        print("")
        dp[n + 1][i][j][k] += p
        dp[n + 1][min(i + 1, a)][j][k] += p
        dp[n + 1][i][min(j + 1, b)][k] += p
        dp[n + 1][min(i + 2, a)][j][k] += p
        dp[n + 1][i][j][min(k + 1, c)] += p
        dp[n + 1][min(i + 1, a)][min(j + 1, b)][k] += p
# print(N, a, b, c)
print(dp)
print(dp[N])
print(dp[N][a])
print(dp[N][a][b])
print(dp[N][a][b][c])
```
