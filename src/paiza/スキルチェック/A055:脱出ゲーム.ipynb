```python
from collections import deque

# 入力処理
def init():
  H, W = list(map(int, input().split()))
  S = [list(input()) for _ in range(H)]
  return H, W, S

# 移動処理
def move(H, W, S):
  # 必要なパラメータ用意
  q = deque([])
  flag = False

  # 初期位置取得
  for i in range(H):
    for j in range(W):
      if S[i][j] == 'S':
        sy, sx = i, j
  q.append((sy,sx))

  # 移動可能な範囲を全探索
  while q:
    cy, cx = q.popleft()
    if (0 <= cy-1 < H) and (S[cy-1][cx] == "."):
      S[cy-1][cx] = "*"
      q.append((cy-1, cx))
    if (0 <= cy+1 < H) and (S[cy+1][cx] == "."):
      S[cy+1][cx] = "*"
      q.append((cy+1, cx))
    if (0 <= cx-1 < W) and (S[cy][cx-1] == "."):
      S[cy][cx-1] = "*"
      q.append((cy, cx-1))
    if (0 <= cx+1 < W) and (S[cy][cx+1] == "."):
      S[cy][cx+1] = "*"
      q.append((cy, cx+1))

    # 境界まできたら，flagをTrueにする
    if not(0 <= cy-1 < H) or \
        not(0 <= cy+1 < H) or \
        not(0 <= cx-1 < W) or \
        not(0 <= cx+1 < W):
      flag = True
      break
  return flag

# 主要関数
def main():
  H, W, S = init()
  flag = move(H, W, S)
  if flag:
    print("YES")
  else:
    print("NO")


if __name__ == '__main__':
  main()
```
