# 1. 裏返せる可能性（縦横）


出力する盤面の情報と石を置くマスの情報が与えられます。

石を置いたマスを '!' に、石を置いたマスと行または列が同じマスを '*' に、それ以外のマスを '.' にした盤面を出力してください。

なお、マスの座標系は左上端のマスの座標を ( y , x ) = ( 0 , 0 ) とし、
下方向が y 座標の正の向き、右方向が x 座標の正の向きとします。

```python
h,w,y,x = list(map(int, input().split()))
s = [["." for _ in range(w)] for _ in range(h)]
```

```python
# h,w,y,x = 3,3,0,0
# h,w,y,x = 10,10,5,5
```

```python
for i in range(h):
  for j in range(w):
    if i == y or j == x:
      s[i][j] = "*"
    if i == y and j == x:
      s[i][j] = "!"

for i in range(len(s)):
  print("".join(map(str, s[i])))
```
